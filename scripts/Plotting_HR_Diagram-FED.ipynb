```python
plt.rcParams['figure.figsize'] = 8,12
d=0
reds = Table["tsneWeird"][valid] > 0
plt.plot(Table["bp-rp"][valid][reds], Table["Solar Luminosity"][valid][reds],
                     '.', color='r', markersize=10)
plt.plot(Table["bp-rp"][valid][~reds], Table["Solar Luminosity"][valid][~reds],
                     '.', color='k', markersize=4, alpha=0.5)
   
plt.xscale('log')
plt.yscale('log')
plt.xlabel("$b'-r'$",fontsize=16)
plt.xticks([0.5, 1, 2, 3], ["0.5", "1", "2", "3"], fontsize=15)
plt.yticks([0.1, 1, 10, 100, 1000], ["0.1", "1", "10", "100", "1000"], fontsize=15)
plt.ylabel(r'Luminosity ($L_\odot$)',fontsize=16)
plt.show()
```
