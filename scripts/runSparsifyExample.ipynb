```python
from sparsify import  sporadicSampling
import numpy as np
import pylab as pl
%pylab inline
```

```python
lcvsKep = np.load("../data/KeplerSampleWErr.npy")
print("shape of lightcurve set", lcvsKep.shape)
pl.plot(lcvsKep[10,0], lcvsKep[10,1])
pl.errorbar(lcvsKep[10,0], lcvsKep[10,1], yerr=lcvsKep[10,2], c='k', 
            fmt='.', alpha=0.5)
pl.title("lcv 10");
pl.xlabel("time (days)")
lcvshape = lcvsKep.shape
```

```python
sampled = sporadicSampling(10, lcvsKep)[1]

pl.plot(lcvsKep[10,0], lcvsKep[10,1])
pl.errorbar(lcvsKep[10,0], lcvsKep[10,1], yerr=lcvsKep[10,2], c='k', 
            fmt='.', alpha=0.5)
pl.errorbar(sampled[10][0], sampled[10][1], yerr=sampled[10][2], fmt='.', lw=3)

pl.title("lcv 10");
pl.xlabel("time (days)")
lcvshape = lcvsKep.shape
```
