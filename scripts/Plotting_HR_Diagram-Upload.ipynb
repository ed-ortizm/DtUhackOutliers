```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

''' Set 'Observables' to the 'HR_Observables.npy' file accompanied to this file, this array contains all 2500 light curves we are using. The Headers can be seen when printing the DataFrame file, to create the HR diagram stick to using the numpy array. Set 'Weirdness' parameter to a metric output from your own code and adjusting the vmin/vmax accordingly.'''

```python
Observables=np.load('../data/HR_Observables.npy')
Weirdness=np.load('/home/dcrake/bin/npy_Files/HR_Diagram/HR_Weirdness.npy') #Your own Weirdness Scores,with corresponding dummy variables.
Weirdness = Weirdness[Weirdness[:,0].argsort()]
```

```python
Data=Observables.T
Table=pd.DataFrame({'Kepler ID':Data[1],'Ra':Data[2], 'Dec':Data[3], 'Object ID':Data[4], 'Object Type':Data[5], 'bp-rp':Data[6], 'Solar Luminosity':Data[7],'Wierdness':Weirdness.T[1]})
Table
```

# Binary HR Diagram

```python
plt.rcParams['figure.figsize'] = 8,12
d=0
for i in range(2500):
#for i in range(2500):

    if float(Observables[i][6])>0 and float(Observables[i][7])>0:
        if Weirdness[i,1]>0.8:
            plt.plot(float(Observables[i][6]),float(Observables[i][7]),'.',color='r',markersize=4)
        else:
            plt.plot(float(Observables[i][6]),float(Observables[i][7]),'.',color='b',markersize=4)
    else:
        d+=1
    if i/100==int:
        print(i)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('bp-rp',fontsize=16)
plt.ylabel('Luminosity, (M_sun)',fontsize=16)
plt.show()
```

# Gradient Color Scale

```python
plt.rcParams['figure.figsize'] = 8,12
cm = plt.cm.get_cmap('BuPu')
for i in range(2500):
    sc = plt.scatter(float(Observables[i][6]),float(Observables[i][7]), s=4, c=Weirdness[i][1], vmin=0.35, vmax=1, cmap=cm)
cb = plt.colorbar(sc)
cb.set_label('URF weirdness score', rotation=90)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r"$b'-r'$",fontsize=16)
plt.ylabel(r'Luminosity ($M_{\odot}$)',fontsize=16)
plt.xlim(xmin=0.25,xmax=4)
plt.ylim(ymin=0.02,ymax=2000)
plt.xticks([0.5, 1, 2, 3], ["0.5", "1", "2", "3"])
plt.show()
#Spectral_r, BuPu,YlGnBu
```

# Gradient Colour Scale - Inverted

```python
plt.rcParams['figure.figsize'] = 8,12
cm = plt.cm.get_cmap('BuPu_r')
for i in range(2500):
    sc = plt.scatter(float(Observables[i][6]),float(Observables[i][7]), s=4, c=Weirdness[i][1], vmin=0.35, vmax=1.1, cmap=cm)
plt.colorbar(sc)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('bp-rp',fontsize=16)
plt.ylabel('Luminosity, (M_sun)',fontsize=16)
plt.xlim(xmin=0.25,xmax=4)
plt.ylim(ymin=0.02,ymax=2000)
plt.show()
#Spectral_r, BuPu,YlGnBu
```
